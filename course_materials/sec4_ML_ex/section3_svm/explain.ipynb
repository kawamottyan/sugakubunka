# 3　サポートベクトルマシン
サポートベクトルマシン(SVM)という分類器について解説します．まずは最も基本的な線形SVMの仕組みを見ていきましょう．

線形SVMには，
 * 線形2値分類の手法(カーネル法を用いることで非線形分類も可能)
 * 外れ値に影響されにくい(サポートベクトルで判別器が決まる)
 
という特徴があります．

同じ線形分類の手法であるロジスティック回帰などとの違いは，データがうまく分離されていることを，マージンと呼ばれる量によって測っていることです．

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

## A　線形SVMの仮定
簡単のため，説明変数を$x_1, x_2$としましょう．

SVMによる2値分類では次のように問題を設定します．

 * 目的変数$y$は1または-1とする(「0または1」，「1または2」などにはしない)．

 * $f(x_1, x_2) = w_1x_1 + w_2x_2 + b$とおいて，$f(x_1, x_2)\geqq 0$のとき$y=1$，$f(x_1, x_2)<0$のとき$y=-1$と判別する．

このように定めておくことで，正しく判別できていることは$y(w_1x_1 + w_2x_2 + b) \geqq 0$と表すことができます．

## B　ハードマージンSVM

SVMではマージン最大化という基準で，訓練データを最もよく判別できるように$w_1, w_2, b$の値を求めます．


 * マージン :  判別境界の周りにデータポイントが存在しない帯状の領域
 * サポートベクトル : マージンの境界上のデータポイント
 * マージンの大きさ :　判別境界からサポートベクトルまでの距離(点と直線の距離)
 
この状況を図示すると次のようになります．

<center><img src="./imgs/margin.png" width=500px></center>

マージンの大きさが最大となるように判別境界を設定するのがSVMです．つまり次のような境界線は引かれません．

これはSVMの仕組みが正則化を含んでいることを意味します．

#### 点と直線の距離の公式
マージンの大きさを計算するために，点と直線の距離の公式を復習しておきましょう．
点$(x_1, x_2)$から直線$w_1x_1 + w_2x_2 + b=0$に下した垂線の長さを，この点と直線間の距離といいます．

この値は，
$$\displaystyle \frac{|w_1x_1 + w_2x_2 + b|}{\sqrt{w_1^2+w_2^2}}$$
で求められることが知られています．

```python
#点の座標と直線の係数を入力
x = np.array([2,4])
w1 = 1
w2 = 1
b = -5


intercept = np.array([[-b/w1, 0],[0, -b/w2]]) #切片を計算


plt.scatter(x[0],x[1])
plt.plot(intercept[:,0],intercept[:,1])

plt.show()
```

```python
#点と直線の距離
np.abs(w1*x[0]+w2*x[1]+b)/np.sqrt(w1**2+w2**2)
```

## B　「マージン最大化」の数式による表現

$w_1,w_2, b$の値に同じ定数$(\not=0)$をかけても直線は変わりません．よって点と直線の距離の計算結果も変わりません．

ところが，$f(w_1, w_2) = w_1x_1 + w_2x_2 + b$の値はかけた定数倍の分変化します．
このことから，必要があれば係数を定数倍変更することによって，

 * データポイントがサポートベクトルであることは$y(w_1 x_1 + w_2x_2 + b) = 1$ と表せる．
 * データポイントがマージンの外側にあることは$y(w_1 x_1 + w_2x_2 + b) \geqq 1$と表せる．

とすることができます．
また，このとき，

 * マージンの大きさは$\displaystyle \frac{1}{\sqrt{w_1^2 +w_2^2}}$と表せる．

ということもわかります．

$\displaystyle \frac{1}{\sqrt{w_1^2 +w_2^2}}$の最大化は$w_1^2 +w_2^2$の最小化と同値であることにも注意しましょう．

以上をまとめると，マージンの最大化は，

条件「各データポイントiについて$y(w_1 x_{1i} + w_2x_{2i} + b) \geqq 1$が満たされる」 の下，$w_1^2 +w_2^2$の最小化

と表現することができます．SVMの学習とは数学的にはこの条件付き最小値問題を解くことに他なりません．

## C　スラック変数とソフトマージンの導入
以上のSVMの基礎となる考え方をハードマージンといいます。しかし、ハードマージンを採用してしまうと線形分離不能な訓練データについては解なしとなってしまいます。線形分離不能な場合でも、不能なりになるべく良い直線を引くための工夫がスラック変数の導入です。スラック変数の導入によってマージン内へのデータの侵入や誤判別を許したマージンをソフトマージンといいます．

#### スラック変数
各データポイントに対して，$\xi = \max\{0, 1-y(w_1x_1+w_2x_2+b)\}$をスラック変数といいます．

これはデータポイントがマージン内に存在しないという条件をどれだけ違反しているかを測っているといえる量です．具体例を通してこのことを確認してみましょう．

**演習**　判別関数が$2x_1 + x_2 -6$であるとき，次の点(yは正解ラベル)について，

 * スラック変数の値はいくつですか．
 * サポートベクトルである点はどれですか．
 * 散布図上の配置とスラック変数の値の関係について考えてみましょう．
 * 各データに対して$y_i(w_1 x_{1i} + w_2x_{2i} + b) \geqq 1-\xi$が満たされていることを確認しましょう．

(1) $(2,4), y = 1$<br> 
(2) $(3,0.5), y = 1$<br>
(3) $(1,3), y = -1$<br>
(4) $(1,4.5), y = -1$


```python
#点の座標と直線の係数を入力
x = np.array([[2,4],[3,0.5],[1,3],[1,4.5]])
y = np.array([1,1,-1,-1])

w1, w2, b = 2, 1, -6
intercept = np.array([[-b/w1, 0],[0, -b/w2]]) #切片を計算

plt.scatter(x[[0,1],0],x[[0,1],1], c="red", label="y=1")
plt.scatter(x[[2,3],0],x[[2,3],1], c="green", label="y=-1")
plt.legend()
plt.plot(intercept[:,0],intercept[:,1])
plt.show()
```

ソフトマージンによるSVMによる学習とは，

 * マージンを大きく = $w_1^2+w_2^2$を小さく
 * 違反を小さく = $\xi$の合計を小さく
 
の二つを同時に最適にする判別関数を求めることです．数学的に正確に述べると，


条件「各データポイントについて$y(w_1 x_1 + w_2x_2 + b) \geqq 1-\xi_i$が満たされる」 の下，$w_1^2 +w_2^2+C\sum \xi_i$の最小化

となります．
ここで，Cは負でない定数で，分析者が設定するハイパーパラメータです．Cが大きいほど違反に厳しくマージンが小さくなります(ハードマージンの状態に近づくともいえる)．

通常，SVMと言ったらソフトマージンを用いたSVMを指します．(以下，この資料でもそうします．)

## D　ヒンジ損失
SVMはヒンジ損失を損失関数としてL2正則化を施した識別機ということができます．このことについて順を追って説明します．

#### ヒンジ関数
$hinge(x) = max(0, a-x)$をヒンジ関数といいます．a = 1の場合のグラフは次のようになります．

```python
def hinge(x, a=1):
    return np.where(a-x > 0, a-x, 0)
```

```python
x = np.linspace(start = -3.0, stop = 3.0, num = 100)
L = hinge(x)
plt.plot(x, L)
plt.show()
```

```python
w1, w2, b = 2, 1, -6
x=np.array([1,3])
y = -1

hinge(y*(w1*x[0]+w2*x[1]+b))
```

ヒンジ損失を導入するとSVMは，

各データポイントに対してヒンジ損失を計算し、その合計と罰則項の和
$$J(w) = \sum_{i=1}^{n}hinge(y_i(w_1x_{1i}+w_2x_{2i}+b))+\lambda(w_1^2+w_2^2)$$
をが最も小さくなるような1次式$f(x;w) = w_1x_1+w_2x_2+b$を予測器として求める手法

ということができます．
$\lambda(w_1^2+w_2^2)$はL2正則化を行っている項でマージン最大化に対応しています．
$\lambda$は正則化の強さを決めるハイパーパラメータで正則化パラメータと呼ばれるものです．上で述べたCとは$C=\frac{1}{\lambda}$の関係があり，このことからCは逆正則化パラメータと呼ばれます．

## E　sklearnによる実装
サポートベクトルマシンはsklearn.svmモジュールのSVCクラスで実行できます．<br>
同じモジュール内にサポートベクトル回帰用のクラスSCRも含まれています．

逆正則化ハイパーパラメータは`C`という引数で指定できます。

```python
# デモデータの生成
# 出力ラベルy = 0のデータを生成
X_train_0 = np.random.multivariate_normal(mean = [1, 1], 
                                          cov = [[1, -0.2], [-0.2, 1]], 
                                          size = 50)
y_train_0 = np.zeros(50)
# 出力ラベルy = 1のデータを生成
X_train_1 = np.random.multivariate_normal(mean = [-1, -1], 
                                          cov = [[1, -0.2], [-0.2, 1]], 
                                          size = 50)
y_train_1 = np.ones(50)
# 2種類のデータを結合する。
X_train = np.vstack((X_train_0, X_train_1))
y_train = np.hstack((y_train_0, y_train_1))
```

```python
# データの可視化
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_train)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
# 入力を標準化する。
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
```

```python
from sklearn.svm import SVC

svc_l= SVC(C = 1.0, kernel = "linear")    # インスタンスを建てる。
svc_l.fit(X_train_sc, y_train)    # fitメソッド：学習させる。
```

```python
# テストデータの入力を生成し、実際に予測させる。
# 格子点上にデータを生成する。
xx, yy = np.meshgrid(
    np.linspace(start = -4.5, stop = 4.5, num = 100), 
    np.linspace(start = -4.5, stop = 4.5, num = 100))
X_test = np.c_[xx.ravel(), yy.ravel()]
# 入力の標準化
X_test_sc = sc.transform(X_test)    # 訓練時の標準化と同じ式を用いる。
# 出力ラベルの予測
y_pred = svc_l.predict(X_test_sc)    # predictメソッド：予測ラベルの計算
y_pred
```

```python
# データの可視化
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_train)
plt.scatter(X_test[:, 0], X_test[:, 1], c = y_pred, alpha = 0.05)
plt.show()
```

#### ハイパーパラメーターCの探索
逆正則化パラメータCの値は適切なものを分析者が選ぶ必要があります．<br>
10分割交差検証で平均正解率を調べてみましょう．

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline

pipe = make_pipeline(StandardScaler(),
                     SVC(C=1.0, kernel="linear")
                    )

scores = cross_val_score(pipe,
                         X = X_train,
                         y = y_train,
                         cv=10)
scores.mean()
```

## F　カーネルSVM
ここまでで述べたSVMは線形2値分類を行う手法でした．さらにカーネル法という工夫を施すことによって，非線形の分類器を作ることもできます．

非線形分類を行うための基本的なアイディアは「データを高次元に送って線形分離し元の空間に引き戻す」というものです．
 
しかし高次元のデータは一般に計算量が膨大で扱いが困難です．そこでカーネル法という工夫が登場します．
これはカーネルトリックとも呼ばれ，「**データを高次元に送ることなしに，**高次元に送ったものと同じ結果を得る」ための巧妙な工夫です．

ここからは
 * 高次元空間での分類が非線形分類につながること
 * カーネル法
の二段階で説明をしていきます．

## G　高次元空間での分類
make_circlesデータセットを用いて，高次元空間での分類が非線形の分類を与えることを見ていきましょう．

これは同心円の内側と外側でクラスが分かれているようなデータです．赤と青を直線で分離することが不可能なことは明らかでしょう．

```python
# どんなデータでデモをするか？
from sklearn.datasets import make_circles

X , y = make_circles(n_samples = 100, factor = 0.5, noise = 0.05)
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], c = "red" )
plt.scatter(X[y == 1, 0], X[y == 1, 1], c = "blue" )
plt.show()
```

ここで，各点ごとに成分の2乗和を計算し，これを3次元目として付け加えます．

すると，付け加えた列は内側ほど小さく，外側ほど大きい値をとることになります．これにより線形分離が可能となります．

```python
(X**2).sum(axis = 1)
```

```python
Xsqsum = (X**2).sum(axis=1)
X_high = np.c_[X, Xsqsum]
X_high.shape
```

```python
from sklearn.svm import SVC

svc_h = SVC(C = 1.0, kernel = "linear")
svc_h.fit(X_high, y)
```

```python
svc_h.predict(X_high)
```

```python
# テストデータの入力を生成し、実際に予測させる。
# 格子点上にデータを生成する。
xx, yy = np.meshgrid(
    np.linspace(start = -1.5, stop = 1.5, num = 101), 
    np.linspace(start = -1.5, stop = 1.5, num = 101))
X_test = np.c_[xx.ravel(), yy.ravel(), xx.ravel()**2 + yy.ravel() **2]

# 出力ラベルの予測
y_pred_test = svc_h.predict(X_test)    # predictメソッド：予測ラベルの計算
y_pred_test
```

```python
# データの可視化
plt.scatter(X_high[:, 0], X_high[:, 1], c = y)
plt.scatter(X_test[:, 0], X_test[:, 1], c = y_pred_test, alpha = 0.05)
plt.show()
```

## H　カーネル法
上のデモでは散布図の観察により1次元加えることによって非線形の分類を行いました．
しかし，実際の非線形分類はより複雑な変数変換を行います．

例えば．2次多項式カーネルと呼ばれるカーネルは2次元の入力を
$$\phi(x,y)=(1,x,y,x^2,xy,y^2)$$
という6次元空間に変換します．データに応じた適切な変換を持つカーネルによって特徴量変換を行うことが出来れば，非線形問題を線形問題に落とすことが出来るのではないか信じるのがカーネル法です．

カーネルには様々な種類がありますが，主に次の2種類がよく使われます．（特徴量の変換関数$\phi$は1次元入力に対する例で書いています。）
* 多項式カーネル：$\phi(x)=(1,x,x^2,\cdots,x^d)$
* Gaussカーネル：$\phi(x)=\left(1,\gamma x,\frac{1}{\sqrt{2!}}\gamma^2 x^2,\frac{1}{\sqrt{3!}}\gamma^3 x^3,\cdots\right)$


## H　カーネルSVMのsklearnによる実装
カーネルは`n_neighbors`引数で指定することができ
* `linear` : 線形カーネル
* `poly`：多項式カーネル。`degree`引数に次数を渡しましょう。
* `rbf`：Gaussカーネル。`gamma`引数に適切な値を渡しましょう。


```python
# デモデータの作成
from sklearn.datasets import make_moons
X_train, y_train = make_moons(n_samples = 100, noise = 0.1, random_state = 2)
```

```python
# データの可視化
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_train)
plt.show()
```

```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler

pipe_k = make_pipeline(StandardScaler(),
                     SVC(C=1.0, kernel="rbf", gamma=0.01)
                    )
pipe_k.fit(X_train, y_train)
```

```python
# テストデータの入力を生成し、実際に予測させる。
# 格子点上にデータを生成する。
xx, yy = np.meshgrid(
    np.linspace(start = -3.1, stop = 3.1, num = 100), 
    np.linspace(start = -2.1, stop = 2.1, num = 100))
X_test = np.c_[xx.ravel(), yy.ravel()]

# 出力ラベルの予測
y_pred = pipe_k.predict(X_test)    # predictメソッド：予測ラベルの計算
y_pred
```

```python
# データの可視化
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_train)
plt.scatter(X_test[:, 0], X_test[:, 1], c = y_pred, alpha = 0.05)
plt.show()
```
