# 4　次元削減
変数の個数が多く複雑なモデルは過学習に陥りやすく、予測精度への貢献が少ない変数は削除するなどの工夫が必要になります。単純に変数を切り捨てるのも方法のひとつですが、ここではより少ない変数で元のデータの情報を表現できるような新しい変数を作り出す方法を紹介します。

このような手法を次元削減といい、その中でも基本的な主成分分析について解説します。

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
from matplotlib import pyplot as plt
```

## A　主成分分析(PCA)
主成分分析は、高次元に散布しているデータに対して、そのデータの散らばりが一番大きい方向を見つける手法です。

```python
# データの作成とプロット．0から99まで縦一列
X0 = np.arange(100).reshape(100,1)
X1 = 3*X0

#X0とX1に期待値０、標準偏差15の正規分布を加える
X_demo = np.append(X0,X1, axis =1) + np.random.normal(loc = 0, scale=15, size=200).reshape(100,2)

#横軸X0, 縦軸X1
plt.scatter(X_demo[:,0], X_demo[:,1])
plt.xlim(-150, 250)
plt.ylim(-50, 350)

plt.show()
```

右上がりの赤線方向が最も散らばりが大きい方向で、第1主成分が定める軸です。それに直交するもう一つの赤線が第2主成分が定める軸です。(縦横比の関係で直交に見えませんが実際は直交しています。)

元のデータが2次元だったことに対応して主成分も2つ存在します。しかし、第1主成分の値だけわかれば元の散布図上のどのあたりの点であるかは概ねわかりそうです。

<center>第一主成分の軸の長さ（分散）＞第二主成分の軸の長さ（分散）</center>

そこで、第2主成分は捨てて第1主成分でデータをとらえることにします。これが次元削減です。

 * 主成分は元の次元分ある。
 
 * 各主成分は直交する。（＝内積が０）
 
 * 分散が大きいほうから第1,第2,第3,...と番号が振られている。
 
 * 分散が大きいほうからいくつか(すべてではない)の主成分のみを採用することで次元削減をする。

```python
from sklearn.decomposition import PCA

# sklearnで実行(詳細はあとで扱います)
PCA = PCA()
PCA.fit(X_demo)
pc = PCA.transform(X_demo)

plt.scatter(X_demo[:,0], X_demo[:,1])

line_tr1  = PCA.inverse_transform(np.c_[np.arange(300)-150,np.zeros(300)])
line_tr2  = PCA.inverse_transform(np.c_[np.zeros(70),np.arange(70)-30])


plt.plot(line_tr1[:,0], line_tr1[:,1], color="red")
plt.plot(line_tr2[:,0], line_tr2[:,1], color="red")

plt.xlim(-150, 250)
plt.ylim(-50, 350)

plt.show()
```

### 数式で表現
主成分分析では$p$個の変数$x_1,x_2,\cdots,x_p$から一次結合$z = a_1x_1+\cdots +a_px_p$の散らばりが最も大きくなるような係数を求めます。

ただし、係数ベクトル$a = (a_1, a_2, \cdots , a_p)$の方向が同じであればノルムが大きいほど散らばりは大きくなります。
最も散らばりが大きい**方向**を求めるために、ノルムが1であるものに制限します。

$a$のノルム $ ||a|| = \sqrt{\sum a_i^2}$

### 試しに一次式を作ってみる
* 一次式の係数を設定してみる。
* 100件のデータ$(x_1,x_2)$に対して$z=a_1x_1+a_2x_2$を求める。
* 100個のzたちの分散を求める。

この分散の値が大きくなるような係数ベクトルa(ただしノルムは1)を探してみましょう。


```python
# a0をいろいろ変えてみましょう．
a0 = np.array([1,-1])
a = a0/np.linalg.norm(a0) #方向を変えずにノルムを1にする
a
```

```python
a@(X_demo.T) 
# @は内積の計算を表す。np.dot(a, X_demo.T)と結果は同じ。@の方が速いらしい。
```

```python
# a0をいろいろ変えてみましょう．
a0 = np.array([20,-123])
a = a0/np.linalg.norm(a0) #方向を変えずにノルムを1にする
(a@(X_demo.T)).var()
```

```python
from sklearn.decomposition import PCA

# sklearnで実行(詳細はあとで扱います)
PCA = PCA()
PCA.fit(X_demo)
pc = PCA.transform(X_demo)


plt.scatter(X_demo[:,0], X_demo[:,1])

#line_tr1  = PCA.inverse_transform(np.c_[np.arange(300)-150,np.zeros(300)])
#line_tr2  = PCA.inverse_transform(np.c_[np.zeros(70),np.arange(70)-30])


plt.plot(line_tr1[:,0], line_tr1[:,1], color="red")
plt.plot(line_tr2[:,0], line_tr2[:,1], color="red")

# a0をいろいろ変えてみましょう．
a0 = np.array([1,-1])
a = a0/np.linalg.norm(a0) #方向を変えずにノルムを1にする

# z = a1 x1 + a2 x2 に対応した 方向を図示
pts = np.linspace(-150, 250, 400)
plt.plot(pts,(a[0]/a[1])*pts+np.mean(X_demo[:,1]),color="blue")

plt.xlim(-150, 250)
plt.ylim(-50, 350)

plt.show()
```

$a$を分散が最大の方向にもっていくには？

* 青い方向の分散＝ 「z = a1 x1 + a2 x2」  の分散

$V(z) = V(a_1 x_1 + a_2 x_2) = a_1^2 V(x_1) + a_2^2 V(x_2) + 2a_1 a_2 \mathrm{Cov}(x_1,x_2)$

↑の式が$a_1,a_2$に対していつ最大になるか

### 係数を求める
分散最大の一次式の係数ベクトル$a$は、**分散共分散行列(データが標準化済みなら相関行列でも同じ)の最大固有値に対応する固有ベクトル**です。このことはLagrangeの未定乗数法によって導かれます．Lagrangeの未定乗数法とは関数の最大・最小を調べる方法の一つ(条件付きの最大・最小を決定できる)です。

```python
#np.covは変数の全ての組み合わせをしてくれる. ddof=1で不偏共分散
S = np.cov(X_demo.T, ddof=1)
S
```

```python
e_val, A = np.linalg.eig(S) #固有値, 固有ベクトルを並べた行列
```

```python
np.linalg.eig(S)
```

```python
S@A[:,0] -e_val[0]*A[:,0] #S×(e_val[1]に対応した)固有ベクトルーS×固有ベクトル = 0
```

```python
# 最大固有値に対応する固有ベクトル
a = A[:,1]
a
```

```python
(A[:,1]@X_demo.T).var(ddof=1)
```

```python
from sklearn.decomposition import PCA

# sklearnで実行(詳細はあとで扱います)
PCA = PCA()
PCA.fit(X_demo)
pc = PCA.transform(X_demo)

# sklearnのPCA
plt.scatter(X_demo[:,0], X_demo[:,1])

line_tr1  = PCA.inverse_transform(np.c_[np.arange(300)-150,np.zeros(300)])
line_tr2  = PCA.inverse_transform(np.c_[np.zeros(70),np.arange(70)-30])


plt.plot(line_tr1[:,0], line_tr1[:,1], color="red")
plt.plot(line_tr2[:,0], line_tr2[:,1], color="red")


# 共分散行列の最大固有値の固有ベクトルから計算
pts = np.linspace(-150, 250, 400)
plt.plot(pts,(a[1]/a[0])*pts,color="orange") #最大固有値の固有ベクトルの線
plt.plot(pts,(A[:,0][1]/A[:,0][0])*pts,color="blue") #２番めに大きい固有値の固有ベクトル方向

plt.xlim(-150, 250)
plt.ylim(-50, 350)

plt.show()
```

### 主成分分析＝分散共分散行列の固有値問題

実はこの分散は対応する固有値と等しくなります。

* 最大分散 ＝ 共分散行列の最大固有値　
* 主成分方向　＝　固有ベクトル
* 主成分方向の分散＝固有値

すぐ後で見るsklearnによる実行ではexplained_variance_で確認することができます。

```python
#共分散行列の固有値
e_val
```

```python
PCA.explained_variance_ #主成分方向の分散の計算
```

**Remark**　なお、主成分の情報をもとに元のデータを**再構成**することもできます。主成分分析では固有値の大きい主成分のみを選択するため、元のデータは完全に復元できるわけではありません。しかし、このような復元はデータから**ノイズ**を除去する操作に対応していると考えられています。このアイディアは主成分分析による異常検知や画像データのクレンジングにに応用されています。

また、主成分分析は**再構成誤差**が最小になるような主成分を求める計算であると解釈できる事も知られています。これを**エッカート・ヤングの定理**といいます。詳しくはMNISTを用いたC節のデモで確認してみましょう。

## B　データの読み込み

これは細胞の画像データから半径など30種類の数値を抽出し、まとめたものです。サイズは569。変数は30個です。

radias(半径)，primeter(周囲),area(面積)など，意味を考えると当たり前に相関するような変数もあります。主成分分析によって似たような変数は主成分得点にまとめられます．

$i$番目の主成分得点： $i$番目の主成分 $z_i = a_{i 1} x_1 + \cdots + a_{i p} x_p = \sum_j^p a_{ij} x_j$ の取る値

```python
df.corr()
```

breast cancerデータの各列はそれぞれデータのスケール(=平均や分散)が大きく異なります。このことはdf.describe()で確認しました。多くの場合、主成分分析の計算をする前に、データのスケールの違いが結果に反映しないように標準化を行います。

```python
# numpy配列に変換
X = df.values
```

```python
#StandardScalerで標準化を行うため, インスタンスを確保。この流れよくやる
scaler = StandardScaler()

scaler.fit(X)
X_scaled = scaler.transform(X) # (データ-平均)/標準偏差
pd.DataFrame(X_scaled)
```

PCAインスタンスを作り、fitメソッドで主成分分析の学習を行います。これにより、主成分を与える一次式が得られます。実際に各データの主成分得点を求める場合はtransformメソッドで計算します。

```python
from sklearn.decomposition import PCA
PCA = PCA()

#主成分分析＝主成分の決定＝分散共分散行列の固有値と固有ベクトルの決定
PCA.fit(X_scaled)

#主成分得点 z_i=（i番目に大きい固有値の固有ベクトルから決まる一次式の値） の値が入っている
X_pca = PCA.transform(X_scaled)

plt.scatter(X_pca[:,0], X_pca[:,1]) #biplot(第1主成分，第2主成分をそれぞれ横軸と縦軸に設定した散布図)をかく
plt.show()
```

```python
X_pca[:10,1] #第一主成分 z_1の式に今回得られているデータの最初の10行を代入したもの
```

```python
X_pca.shape #主成分は30個できてるけどほとんど無駄（＝その方向分散が小さい＝その方向にほどんど動きがない）→次元削減してしまう
```

### 主成分の選択
分散が大きいほうからいくつかの主成分を採用し、次元削減を行いましょう。第何主成分まで採用するかが問題になりますが、これは主成分の寄与率という値を用いて判断します。寄与率は主成分の分散を元の変数の分散の合計で割った値です。explained_variance_ratio_で求めることができます。主成分の分散について、次のことが知られています。

 * 主成分の分散は共分散行列の固有値
 
 * 主成分の分散の合計 = 元の変数の分散の合計（非自明）
 
したがって寄与率の合計は1となり、その固有値が元のデータの情報をどれだけ持っているかを表していると考えられます。

###  すなわち 主成分の分散の合計＝元の変数の分散の合計＝固有値の合計
$z_i$の寄与率＝$i$番目に大きい固有値 $\div$ 固有値の合計 = $z_i$の分散 $\div$ 全分散の合計

```python
#第一主成分から寄与率を並べる。第一主成分が一番分散＝固有値大きいので寄与率も最大
PCA.explained_variance_ratio_
```

```python
#寄与率の途中和. 例えば第二主成分までで63%, 第７成分までで91%説明できてることがわかる.
np.cumsum(PCA.explained_variance_ratio_)
```

```python
1/30 #30個の成分が均等に寄与しているなら寄与率これくらい.
```

```python
# 画像を再構成した結果
#第何主成分まで使うか
n_components = [4, 8, 12, 16, 20, 24]

#画像サイズの指定
plt.figure(figsize = (10, 60))

#4～24までのfor文. enumerateを使うとループ回数, 元という順で取れる.今回は[0,4],[1,8],[2,12],[3,16],[4,20],[5,24]ととる
for loc, n_component in enumerate(n_components):
    
    pcas = np.zeros(64) #64個0が並んだものをとり、
    
    pcas[0:n_component] = x[0:n_component] #n_component番目までをxの成分にする.つまり, n_component番目以降は全部0（次元削減で情報消える）
    
    pcas = pcas.reshape(1, 64) # 1 × 64 ベクトルに reshape
    
    #PCA.invere_transformで主成分軸ではない元の軸の世界に, scaler.inverse_transformで標準化も戻す
    img = scaler.inverse_transform(PCA.inverse_transform(pcas)).reshape(8, 8)
    
    #画像を小さくして並べる処理.
    plt.subplot(1, 6, loc+1)
    plt.imshow(img, cmap = "gray")
    
plt.show()
```
