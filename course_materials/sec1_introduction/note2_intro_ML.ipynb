## 2.2 機械学習で目指すもの


先ほど例に挙げたタスクには共通点があります。それは**「入力」から適切な「出力」を返す仕組み<sup>[1](#note1)</sup>** を作ることを目標にしている点です。特に機械学習では、ルールベースシステムなどのように入力から出力を計算するためのルールを明示的に準備できなくても、**データ**からその仕組みをつくることを目指します。

例えば光学文字認識では
* 入力 : 手書き文字画像
* 出力 : 対応する文字コード

を想定しています。ルールベースシステムでこの問題に挑むならば、文字画像が全体的に丸みを帯びていれば「0」というようなルールをあらかじめ明示的に準備して、出力を計算する仕組みを作ることになります。一方、機械学習では手書き文字画像とその対応する文字コードをセットにしたデータからこのような仕組みを作ることを目指します。

**注釈**　<small id="note1">1. 数学では「関数」とか「写像」という名前で親しまれています。</small>

##  2.3 機械学習の魅力的なところ

機械学習の魅力的な点として、例えば次の2点が挙げられるでしょう。
* 自動化
* 複雑なルールをもつタスクにおける優位性

「自動化」は機械学習に限らずルールベースシステムにも当てはまりますが、それでも特筆すべき話です。例えばweb広告の配信では、配信枠を0.1秒もの速さで競り合っています。このような迅速なタスクに対して人は無力ですが、機械学習によって「枠の情報」から適切な「入札額」を決める仕組みを事前に作ることが出来れば対応できます。

また機械学習はルールベースシステムなどに比べて「複雑なルールをもつタスク」において優位です。例えば光学文字認識では、すべての手書き文字画像を文字コードにエンコードするためのルールは大変複雑そうに見えます。ルールの複雑さの分だけ、準備にも膨大なコストがかかるでしょう。要するにルールベースシステムには不向きです。機械学習では、このようなタスクに対してとても上手いアプローチをとります。

## 2.4 機械学習を通して何を学ぶのか？


さて、私たちは機械学習を通して何を学ぶのでしょうか。**目の前に複雑な関係性があったとき、そこからどのようにして所望のルールを取り出すか**を学びます。例えば光学文字認識であれば「手書き文字画像」を「文字コード」にエンコードするルールであり、異音検知であれば「音の波形」から「その音の異常度」を見積もるルールです。


そこには地味で大変な作業や難しい数学が出てくることもありますが、機械学習の基本を上手に使いこなすことができるようになれば、きっとそれらも少しずつ敷居の低いものになっていくことでしょう。
