```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.random.multivariate_normal(mean=[1,1],
                                       cov=[[1,-0.4],[-0.4,1]],
                                       size = 100)
y_train = np.random.binomial(n=1,p=1/(1+np.exp(-2.0*X_train[:,0]-2.0*X_train[:,1]+4.0)))
#binomialは二項分布->0か1の2通りが返ってくる
#X_trainの座標の具合でp(確率)を決めている
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
X_train_ss = scale.fit_transform(X_train)
#標準化でSCALEを統一しないと、説明変数が複数あるとき正則化項の影響の具合がかわってしまう
```

```python
# a = n/C
# Cは逆正規化パラメータ
lr = LogisticRegression(C=100.0, penalty='l2')
lr.fit(X_train_ss,y_train)
```

```python
lr.predict(X_train_ss)
```

```python
(y_train == lr.predict(X_train_ss)).mean()
```

```python
lr.predict_proba(X_train_ss)
```

```python
#予測範囲を散布図に可視化させる。
axls = np.linspace(-2.0,4.0,
                   num = 100)
xx, yy = np.meshgrid(axls,axls)
X_test = np.c_[xx.ravel(),yy.ravel()]
X_test_ss = scale.transform(X_test)
y_prod = lr.predict_proba(X_test_ss)[:,1]
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.scatter(X_test[:,0],X_test[:,1],c=y_prod,alpha=0.05)
plt.show()
```

```python
from sklearn.datasets import load_digits
X,y = load_digits().data, load_digits().target
```

```python
X[0,:].reshape(8,8)
```

```python
X.shape,y.shape
```

```python
plt.imshow(X[0,:].reshape(8,8),cmap="gray")
plt.show
```

```python
index = np.where((y==7)|(y==9))
index
```

```python
X,y = X[index],y[index]
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
lr = LogisticRegression(C=100.0)
lr.fit(X_train,y_train)

pred_train = lr.predict(X_train)
pred_test = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_true = y_train, y_pred = pred_train),accuracy_score(y_true=y_test,y_pred=pred_test)
```

```python
confusion_matrix(y_true = y_train, y_pred = pred_train),confusion_matrix(y_true=y_test,y_pred=pred_test)
```
