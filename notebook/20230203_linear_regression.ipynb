```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#データの作成
X_train = np.linspace(-1,1,num=50)
X_train
```

```python
y_train = 1.0 + 2 * X_train + np.random.normal(loc= 0.0, scale=1.0, size=50)
y_train
```

```python
plt.scatter(X_train,y_train)
plt.show()
```

```python
X_train = X_train.reshape(50,1)
X_train
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train) #学習
```

```python
y_pred = LR.predict(X_train) #予測
y_pred
```

```python
y_train
```

```python
#回帰係数
LR.coef_ , LR.intercept_
```

```python
X_test = np.linspace(-1,1,10000)
y_test = LR.predict(X_test.reshape(10000,1))
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_test,y_test) #予測値の折れ線グラフ
plt.show()
```

```python
#多重共線性
X_train = np.random.multivariate_normal(mean = [0, 0],
                                        cov = [[1.0, 0.9],
                                               [0.9, 1.0]],
                                        size = 100)
y_train = 1.0+X_train[:,0] + np.random.normal(loc=0.0,scale=1.0,size=100)

LR.fit(X_train,y_train)
LR.intercept_,LR.coef_
```
