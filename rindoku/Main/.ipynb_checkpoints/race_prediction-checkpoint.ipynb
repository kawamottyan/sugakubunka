```python
import numpy as np
import pandas as pd

import seaborn as sns
from matplotlib import pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
race_df = pd.read_csv("../Scraping/csv/cleaned_race_data.csv")
horse_df = pd.read_csv("../Scraping/csv/cleaned_horse_data.csv")
```

# EDA

```python
race_df.corr()
```

# PCA

```python
plt.figure(figsize=(15,15)) 
sns.heatmap(race_df.corr(), square=True, annot=True)
plt.show()
```

```python
plt.figure(figsize=(15,15)) 
sns.heatmap(horse_df.corr(), square=True, annot=True)
plt.show()
```

```python
#LabelEncoder
for column in ['race_title','weather','ground_status','date','where_racecourse','is_left_right_straight']:
    le = LabelEncoder()
    le.fit(race_df[column])
    race_df[column] = le.transform(race_df[column])
```

```python
# numpy配列に変換
X_race = race_df.values
X_horse = horse_df.values
```

```python
#StandardScalerで標準化を行うため, インスタンスを確保。この流れよくやる
scaler = StandardScaler()

scaler.fit(X_race)
X_race_scaled = scaler.transform(X_race) # (データ-平均)/標準偏差b
```

```python
PCA = PCA()

#主成分分析＝主成分の決定＝分散共分散行列の固有値と固有ベクトルの決定
PCA.fit(X_race_scaled)

#主成分得点 z_i=（i番目に大きい固有値の固有ベクトルから決まる一次式の値） の値が入っている
X_pca = PCA.transform(X_race_scaled)

plt.scatter(X_pca[:,0], X_pca[:,1]) #biplot(第1主成分，第2主成分をそれぞれ横軸と縦軸に設定した散布図)をかく
plt.show()
```

```python
PCA.explained_variance_ratio_
```

# モデル作成

```python
y = horse_df['rank'].values
X = horse_df.drop('rank', axis=1).values
```

```python
X_train, X_test, y_train, y_test =\
            train_test_split(X, y, 
                             test_size=0.2, 
                             random_state=1,
                             stratify=y)
```

```python
#弱学習器の作成
tree = DecisionTreeClassifier(criterion='entropy', 
                              max_depth=1,
                              random_state=1)
#アダブーストの作成
ada = AdaBoostClassifier(base_estimator=tree,
                         n_estimators=500, 
                         learning_rate=0.1, 
                         random_state=1)
```

```python
#弱学習器の予測
tree = tree.fit(X_train, y_train)
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
#弱学習器のスコア
tree_train = accuracy_score(y_train, y_train_pred)
tree_test = accuracy_score(y_test, y_test_pred)
print('Decision tree train/test accuracies %.3f/%.3f'
      % (tree_train, tree_test))

#アダブーストの予測
ada = ada.fit(X_train, y_train)
y_train_pred = ada.predict(X_train)
y_test_pred = ada.predict(X_test)
#アダブーストのスコア
ada_train = accuracy_score(y_train, y_train_pred) 
ada_test = accuracy_score(y_test, y_test_pred) 
print('AdaBoost train/test accuracies %.3f/%.3f'
      % (ada_train, ada_test))
```

```python
fi = ada.feature_importances_  
print('Feature Importances:')
for i, feat in enumerate(horse_df.columns[]):
    print('\t{0:20s} : {1:>.6f}'.format(feat, fi[i]))
```

# ハイパーパラメータ探索

```python
param_grid = {
    'n_estimators': [100, 300, 500, 800, 1000],
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.5]
}

ada = AdaBoostClassifier(base_estimator=tree,
                         random_state=1)
grid_search = GridSearchCV(ada, param_grid, cv=5)
gs = grid_search.fit(X_train, y_train)

y_train_pred = gs.predict(X_train)
y_test_pred = gs.predict(X_test)

gs_train = accuracy_score(y_train, y_train_pred) 
gs_test = accuracy_score(y_test, y_test_pred) 
print('GridSearch train/test accuracies %.3f/%.3f'
      % (gs_train, gs_test))
```

```python
best_parameters = grid_search.best_params_
best_parameters
```
