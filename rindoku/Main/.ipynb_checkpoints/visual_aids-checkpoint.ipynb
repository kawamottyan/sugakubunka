```python
import numpy as np
import matplotlib.pyplot as plt
```

# Alphaのグラフ

```python
#x軸
error_rate = np.linspace(0, 1, 100)

#y軸
#alphaの定義
alpha = 1/2 *np.log((1 - error_rate)/ error_rate)
```

```python
#プロット
plt.plot(error_rate, alpha)
plt.xlabel("error_rate")
plt.ylabel("alpha")
plt.show()
```

# 重み

### 不正解

```python
#x軸
alpha = np.linspace(-2, 2, 100)

#y軸
#重みの定義
wight = np.exp(alpha)
```

```python
plt.plot(alpha, wight)
plt.xlabel("alpha")
plt.ylabel("wight")
plt.show()
```

### 正解

```python
#x軸
alpha = np.linspace(-2, 2, 100)

#y軸
#重みの定義
wight = np.exp(-alpha)
```

```python
plt.plot(alpha, wight)
plt.xlabel("alpha")
plt.ylabel("wight")
plt.show()
```
